### Split
- převede string na list

```python
words = text.split()

print("\n=== LIST SLOV ===")
print(words)

print(f"\nPočet prvků v listu: {len(words)}")

# Přístup k prvkům listu přes index (čísluje se od 0!)
print(f"\nPrvní slovo  (index 0):  '{words[0]}'")
print(f"Druhé slovo  (index 1):  '{words[1]}'")
print(f"Poslední sl. (index -1): '{words[-1]}'")

# Slice — výřez z listu
print(f"\nPrvních 5 slov (slice [0:5]): {words[0:5]}")
```

```python
split_a = text.split('a')
print(f"\n=== SPLIT PODLE 'a' ===")

print(split_a)
```

### For loop
- iterace přes prvky objektu
- string
- list
- tuple
- dictionary
- set
- range


```python
for word in words:
    print(word)
```

```python
# For loop s indexem (enumerate nám dá číslo i hodnotu najednou)
print("\n=== FOR LOOP s enumerate: slovo + jeho pozice ===")

for index, word in enumerate(words):
    print(f"  pozice {index}: {word}")
```

```python
# počítání prvků přes for loop
print("\n=== POČÍTÁNÍ PRVKŮ S FOR LOOP ===")
counter = 0
for word in words:
    counter += 1

print(f"Počet slov spočítaný for loopem: {counter}")
```

### Vytvoření listu pomocí for loop

```python
lower_words = []

for word in words:
    lower_word = word.lower()
    lower_words.append(lower_word)   # .append() přidá prvek na konec listu

print("\n=== MALÁ PÍSMENA (for loop) ===")
print(lower_words)
```

### List comprehension

```python
# Totéž co výše, ale na jeden řádek:
lower_words_v2 = [word.lower() for word in words]

print("\n=== MALÁ PÍSMENA (list comprehension) ===")
print(lower_words_v2)

# Jsou oba výsledky stejné?
print(f"\nShodují se oba přístupy? {lower_words == lower_words_v2}")
print(1==1)

# jedno = definuje proměnnou, dvě == porovnávají hodnoty (vrací True/False)
```

### If-else

```python
# Chceme vypsat každé slovo, ale označit ta delší než 6 znaků

for word in lower_words:
    if len(word) > 6:
        print(f"  DLOUHÉ:  {word}  ({len(word)} znaků)")
    else:
        print(f"  krátké:  {word}  ({len(word)} znaků)")
```

```python
# Chceme ověřit, zda se má slovo nějaký znak (např. 'a') a podle toho ho označit

print(f'Je a to stejné co A? {"a" == "A"} - Ne, protože python je case-sensitive (rozlišuje malá a velká písmena)')

words_with_a = []
words_without_a = []
words_starting_with_a = []
words_ending_with_a = []

for word in lower_words:
    if 'a' in word:
        words_with_a.append(word)
    else:
        words_without_a.append(word)

    if word.startswith('a'):
        words_starting_with_a.append(word)

    if word.endswith('a'):
        words_ending_with_a.append(word)

print(f"\nSlova obsahující 'a': {words_with_a}")
print(f"Slova neobsahující 'a': {words_without_a}")
print(f"Slova začínající na 'a': {words_starting_with_a}")
print(f"Slova končící na 'a': {words_ending_with_a}")
```

```python
# čištění duplicit z listu pomocí set (množina, která neobsahuje duplicitní prvky)

list_without_duplicates = list(set(words_ending_with_a))
print(f"\nSlova bez duplicit: {list_without_duplicates}")
```

```python
# If-elif-else pro více podmínek - pozor, splní se jen první podmínka, která je pravdivá, a zbytek se už nevyhodnocuje! (důležité pro hierarchii podmínek)

for word in lower_words:
    if word.startswith('a') and word.endswith('a'):
        print(f"Slovo začínající i končící na 'a': {word}")
    elif word.startswith('a'):
        print(f"Slovo začínající na 'a': {word}")
    elif word.endswith('a'):
        print(f"Slovo končící na 'a': {word}")
    elif 'a' in word:
        print(f"Slovo obsahující 'a' někde uprostřed: {word}")
    else:
        print(f"Slovo neobsahující 'a' na začátku ani na konci: {word}")
```

### If-else pro filtrování

```python
stopwords = ["a", "se", "je", "v", "na", "z", "o", "k", "i", "nebo", "s", "u", "nad", "pod",
             "jak", "jsou", "tedy", "tedy", "jej", "mu", "po",
             "ke", "bez", "ale", "ten", "každý", "každou"]


content_words = []

for word in lower_words:
    if word in stopwords:
        print(f"  přeskakuji stopword: '{word}'")
    else:
        content_words.append(word)
        print(f"  přidávám:            '{word}'")

print(f"\nContent words: {content_words}")
print(f"Původní počet slov: {len(lower_words)}")
print(f"Po filtrování:      {len(content_words)}")
```
