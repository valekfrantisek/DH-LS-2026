### If-else

```python
# Chceme vypsat každé slovo, ale označit ta delší než 6 znaků

for word in lower_words:
    if len(word) > 6:
        print(f"  DLOUHÉ:  {word}  ({len(word)} znaků)")
    elif len(word) == 6:
        print(f"  střední: {word}  ({len(word)} znaků)")
    else:
        continue
        # print(f"  krátké:  {word}  ({len(word)} znaků)")
```

```python
# Chceme ověřit, zda se má slovo nějaký znak (např. 'a') a podle toho ho označit

print(f'Pozor!! Je "a" to stejné co "A"? {"a" == "A"} - Ne, protože python je case-sensitive (rozlišuje malá a velká písmena)')

words_with_a = []
words_without_a = []
words_starting_with_a = []
words_ending_with_a = []

for word in lower_words:
    if 'a' in word:
        words_with_a.append(word)
    else:
        words_without_a.append(word)

    if word.startswith('a'):
        words_starting_with_a.append(word)

    if word.endswith('a'):
        words_ending_with_a.append(word)

print(f"\nSlova obsahující 'a': {words_with_a}")
print(f"Slova neobsahující 'a': {words_without_a}")
print(f"Slova začínající na 'a': {words_starting_with_a}")
print(f"Slova končící na 'a': {words_ending_with_a}")
```

```python
# čištění duplicit z listu pomocí set (množina, která neobsahuje duplicitní prvky)
print(f"\nSlova končící na 'a' (s duplicitami): {words_ending_with_a}")

list_without_duplicates = list(set(words_ending_with_a))
print(f"\nSlova bez duplicit: {list_without_duplicates}")
```

```python
# If-elif-else pro více podmínek - pozor, splní se jen první podmínka, která je pravdivá, a zbytek se už nevyhodnocuje! (důležité pro hierarchii podmínek)

for word in lower_words:
    if word.startswith('a') and word.endswith('a'):
        print(f"Slovo začínající i končící na 'a': {word}")
    elif word.startswith('a'):
        print(f"Slovo začínající na 'a': {word}")
    elif word.endswith('a'):
        print(f"Slovo končící na 'a': {word}")
    elif 'a' in word:
        print(f"Slovo obsahující 'a' někde uprostřed: {word}")
    else:
        print(f"Slovo neobsahující 'a' na začátku ani na konci: {word}")
```

### If-else pro filtrování (hodina 3. 3. 2025)

```python
stopwords = ["a", "se", "je", "v", "na", "z", "o", "k", "i", "nebo", "s", "u", "nad", "pod",
             "jak", "jsou", "tedy", "tedy", "jej", "mu", "po",
             "ke", "bez", "ale", "ten", "každý", "každou"]


content_words = []

for word in lower_words:
    if word in stopwords:
        print(f"  přeskakuji stopword: \"{word}\"")
        continue
    else:
        content_words.append(word)
        print(f"  přidávám:            \"{word}\"")

print(f"\nContent words: {content_words}")
print(f"Původní počet slov: {len(lower_words)}")
print(f"Po filtrování:      {len(content_words)}")
```

# Funkce

```python
""" Funkce můžeme definovat pro opakující se úkoly, které chceme znovu použít. Funkce může mít parametry (vstupy) a může vracet výstup (return). """

def counter(a, b):
    result = a + b
    return result

# def simple_function(a, b):
#     return a + b
```

```python
sum_result = counter(50, 101)
print(f"\nVýsledek sčítání: {sum_result}")
```

```python
""" Funkcním můžeme zadat parametry a určit jim, jaké hodnoty mají přijímat (např. text a seznam stopwords) a pak v těle funkce provést nějakou operaci (např. spočítat obsahová slova). """

def count_content_words(text:str, stopwords:list): 
    words = text.split()
    content_words = []
    for word in words:
        if word.lower() not in stopwords:
            content_words.append(word)
            
    return len(content_words)

""" Stejnou funkcionalitu můžeme zapsat i pomocí list comprehension, což je často kratší a přehlednější způsob, jak vytvořit nový list na základě existujícího. """

def count_content_words_v2(text:str, stopwords:list):            
    return len([word.lower() for word in text.split() if word.lower() not in stopwords])
```

```python
stopwords = ["a", "se", "je", "v", "na", "z", "o", "k", "i", "nebo", "s", "u", "nad", "pod",
             "jak", "jsou", "tedy", "tedy", "jej", "mu", "po",
             "ke", "bez", "ale", "ten", "každý", "každou"]

lower_text = text.lower()

count_content = count_content_words(lower_text, stopwords)
print(f"\nPočet content words: {count_content}")
print(f"Počet slov celkem: {len(lower_text.split())}")

count_content = count_content_words_v2(text=lower_text, stopwords=stopwords)
print(f"\nPočet content words: {count_content}")
print(f"Počet slov celkem: {len(lower_text.split())}")
```

```python
count_stop = count_content_words(lower_text, stopwords)
print(f"\nPočet content words: {count_stop}")
```

```python
""" Můžeme také nastavit výchozí hodnotu pro parametr stopwords, takže pokud funkci zavoláme bez zadání stopwords, použije se náš defaultní seznam. """

def count_content_words_default(text:str, stopwords:list=None):
    if stopwords is None:
        stopwords = ["a", "se", "je", "v", "na", "z", "o", "k", "i", "nebo", "s", "u", "nad", "pod",
             "jak", "jsou", "tedy", "tedy", "jej", "mu", "po",
             "ke", "bez", "ale", "ten", "každý", "každou"]
        
    words = text.split()
    content_words = []
    for word in words:
        if word.lower() not in stopwords:
            content_words.append(word)
            
    return len(content_words)
```

```python
""" Nyní můžeme zavolat funkci bez zadání stopwords, protože máme nastavenou výchozí hodnotu. """
count_stop = count_content_words_default(lower_text)
print(f"\nPočet content words: {count_stop}")
```

```python
""" Nebo můžeme zadat vlastní seznam stopwords, který se bude lišit od výchozího. """
count_stop = count_content_words_default(lower_text, stopwords=['a', 'se', 'je'])
print(f"\nPočet content words: {count_stop}")
```

```python
bible_files = os.listdir(BIBLE_PATH)
print(bible_files)
```

```python
BKR_books = {}
BKR_verses = {}

for bf in bible_files:
    with open(os.path.join(BIBLE_PATH, bf), 'r', encoding='utf-8') as f:
        f_data = f.read()
    
    bible_dict = eval(f_data)  # převod stringu na slovník ("nebezpečné", ale pro tento účel to bude v pořádku)
    BKR_books[bf] = bible_dict

    for key, value in bible_dict.items():
        BKR_verses[key] = value

print(len(BKR_books))
print(len(BKR_verses))
  
```

```python
def print_book_chap_verses(book_id, chapter:int):
    for key, value in BKR_verses.items():
        if key.startswith(f'{book_id} {chapter}:'):
            print(f'{key}, {value}')
```

```python
print_book_chap_verses('Z', 1)
```

```python
def content_words_default(text, stopwords=None):
    if stopwords is None:
            stopwords = ["a", "se", "je", "v", "na", "z", "o", "k", "i", "nebo", "s", "u", "nad", "pod",
                "jak", "jsou", "tedy", "tedy", "jej", "mu", "po",
                "ke", "bez", "ale", "ten", "každý", "každou",
                ".", "!", "?", ",", ";", ":", "-", "—", "(", ")", "[", "]", "{", "}", '"', "'", "/", '\\']
        
    words = nltk.word_tokenize(text)
    content_words = []
    for word in words:
        if word.lower() not in stopwords:
            content_words.append(word)
            
    return content_words
```

```python
words_BKR = []

for verse in BKR_verses.values():
    words_BKR.extend(content_words_default(verse))  # .extend() přidá všechny prvky z daného listu do našeho listu
```

```python
freq_dist_BKR = FreqDist(words_BKR)
freq_dist_BKR.most_common(20)
```

```python
new_stops = []
for word, frequency in freq_dist_BKR.most_common(50):
    print(f"{word}: {frequency}")
    new_stops.append(word)

print(new_stops)
```
