```python
# Čísla
cislo_int = 42          # int (celé číslo)
cislo_float = 3.14      # float (desetinné číslo)
cislo_complex = 2 + 3j  # complex (komplexní číslo)

# Text
text = "Ahoj, světe!"   # str (řetězec)
znak = 'Ahoj "zajksjdkjsd"'

# Pravdivostní hodnota
pravda = True           # bool
nepravda = False

# Nic
nic = None              # NoneType

# Seznam (měnitelný, seřazený)
seznam = [1, 2, 3, "čtyři", True]
seznam.append('5')

# N-tice (neměnitelná, seřazená)
ntice = (10, 20, 30)

# Slovník (klíč → hodnota)
slovnik = {
    "jméno": "Karel",
    "jméno": 'jan',
    "věk": 30,
    "programátor": True
}

# Množina (unikátní hodnoty, neseřazená)
mnozina = {1, 2, 3, 2, 1}  # → {1, 2, 3}
```

```python
class Auto:
    def __init__(self, znacka, rok):
        self.znacka = znacka
        self.rok = rok

    def info(self):
        return f"{self.znacka} ({self.rok})"

moje_auto = Auto("Škoda", 2020)
print(moje_auto.info())  # → Škoda (2020)
```

```python
def pozdrav(jmeno):
    return f"Ahoj, {jmeno}!"

# Funkci lze přiřadit do proměnné nebo předat jako argument
muj_pozdrav = pozdrav
print(muj_pozdrav("Petr"))  # → Ahoj, Petr!
```
