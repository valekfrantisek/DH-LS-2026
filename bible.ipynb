```python
""" Importování potřebných knihoven. """

import os
import re
import nltk
from collections import defaultdict
```

```python
""" Nadefinování cest k souborům a adresářům. """

ROOT = os.getcwd()
print(f"Current working directory: {ROOT}")
print(type(ROOT))

BIBLE_PATH = os.path.join(ROOT, 'bible_files')

EXAMPLE_PATH = os.path.join(ROOT, 'example_corpus.txt')
```

```python
""" Načítaní všech biblických veršů ze souborů do slovníku. """

all_verses = {}

for b_file in os.listdir(BIBLE_PATH):
    with open(os.path.join(BIBLE_PATH, b_file), 'r', encoding='utf-8') as f:
        b_data = f.read()
        
        book_dict = eval(b_data)
        all_verses.update(book_dict)

        print(len(book_dict), "verses read.... updated to ", len(all_verses), "verses in total.")
```

```python
""" Vytvoření inverzního indexu pro rychlé vyhledávání veršů podle slov. """

words_to_verses = defaultdict(list)

stop_words = ["a", "se", "je", "v", "na", "z", "o", "k", "i", "nebo", "s", "u", "nad", "pod",
             "jak", "jsou", "tedy", "tedy", "jej", "mu", "po", 'od', 'do', 'pro', 'bez', 'ale', 'ten', 'každý', 'každou',
             "ke", "bez", "ale", "ten", "každý", "každou", 'co', 'což', 'kdo', 'který', 'která', 'které', 'kterým', 'kterými', 'kterého', 'kterému', 'kterou']

for verse_id, verse_text in all_verses.items():
    words = re.findall(r'\b\w+\b', verse_text.lower())
    for word in words:
        if word not in stop_words:
            words_to_verses[word].append(verse_id)

print(f"Total unique words: {len(words_to_verses)}")
```

```python
""" Ukázka několika slov a veršů, ve kterých se vyskytují. """
for key, value in list(words_to_verses.items())[:10]:  # Print first 10 entries
    print(f"Word: '{key}' appears in verses: {value[:5]}...")  # Show first 5 verse IDs for each word
```

```python
""" Načítaní textu, který chceme dotazovat. """

with open(EXAMPLE_PATH, 'r', encoding='utf-8') as f:
    data_to_query = f.read()

print(len(data_to_query), "characters read from example corpus.")
```

```python
""" Rozdělení textu na věty a výběr jedné věty pro dotazování. """

sents_in_corpus = nltk.sent_tokenize(data_to_query)

sent_to_query = sents_in_corpus[2]
print(sent_to_query)
```

```python
""" Extrakce slov z věty pro dotazování. (Extrahujeme slova pro případné porovnávání s verši. Extrakce probíhá stejně jako při vytváření inverzního indexu.) """

words_in_sent = re.findall(r'\b\w+\b', sent_to_query.lower())
print(f"Words in sentence: {words_in_sent}")
```

```python
""" Vyhledání veršů, které obsahují slova z věty pro dotazování, a přiřazení skóre podle počtu shodných slov. """

score_dict = defaultdict(int)

for word in words_in_sent:
    if word in words_to_verses:
        for verse_id in words_to_verses[word]:
            score_dict[verse_id] += 1
```

```python
""" Zobrazení top 5 veršů s nejvyšším skóre. Jen pro ukázku. """

print("Top 5 verses by score:")
for verse_id, score in sorted(score_dict.items(), key=lambda x: x[1], reverse=True)[:5]:
    print(f"Verse ID: {verse_id}, Score: {score}")

print(len(score_dict), "verses scored based on query sentence." )
```

```python
""" Zobrazení počtu veršů, které mají skóre 3 nebo vyšší. V rámci těchto veršů následně budeme hledat ten, který je nejvíce podobný dotazované větě. """

verses_with_score_greater_than_3 = {vid: sc for vid, sc in score_dict.items() if sc >= 3}
print(len(verses_with_score_greater_than_3), "verses have a score of 3 or higher.")
```

```python
""" Funkce pro rozdělení verše na části, které budou porovnávány s dotazovanou větou. Pro jednoduchost zde používáme rozdělení podle čárek a teček, ale v praxi by bylo vhodné použít sofistikovanější metodu. """

def split_verse(verse_id):
    verse_text = all_verses[verse_id]
    verse_sentences = nltk.sent_tokenize(verse_text)

    verse_parts = []
    for sent in verse_sentences:
        sent_pats = sent.split(',')

        for sent_pat in sent_pats:
            if len(sent_pat.split()) > 2:
                verse_parts.append(sent_pat.strip())
    
    return verse_parts
```

```python
""" Zobrazení částí veršů (segmentů), které mají skóre 3 nebo vyšší. """

for verse_id in verses_with_score_greater_than_3:
    print(verse_id, split_verse(verse_id))
```

```python
""" Hledání přesné shody mezi segmenty veršů s vysokým skóre a dotazovanou větou. Pokud se nějaký segment přesně shoduje s částí dotazované věty, vypíšeme tento segment a ID verše. """

for verse_id in verses_with_score_greater_than_3:
    for segment in split_verse(verse_id):
        if segment.lower() in sent_to_query.lower():
            print(f"Exact match found in verse {verse_id}: '{segment}'")
```

```python
import nltk

# 1. Vyčistíme si celou Bibli jen JEDNOU dopředu (tohle zabere asi 2 vteřiny)
clean_BKR_verses = {}
for verse_id, verse_text in all_verses.items():
    # Rychlé vyčištění: převedeme na malá písmena a odstraníme interpunkci
    clean_text = "".join([c if c.isalpha() or c.isspace() else " " for c in verse_text.lower()])
    clean_BKR_verses[verse_id] = " ".join(clean_text.split())

# 2. Samotná funkce pro hledání
def find_citation_in_sentence(sentence:str):
    verse_id_hit = False
    
    words = nltk.word_tokenize(sentence.lower())
    clean_words = [w for w in words if w.isalpha()]
    
    okno = 5 # Počet slov pro shodu
    
    if len(clean_words) >= okno:
        for i in range(len(clean_words) - okno + 1):
            fraze = " ".join(clean_words[i:i+okno])
            
            # Hledáme už v těch předpřipravených, vyčištěných verších - BLESKOVĚ
            for verse_id, clean_verse in clean_BKR_verses.items():
                if fraze in clean_verse:
                    verse_id_hit = verse_id
                    return verse_id_hit
                    
    return verse_id_hit
```
